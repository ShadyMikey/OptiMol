```python
# extracting atoms
from ccdc.io import MoleculeReader
import pandas as pd
import re
from gemmi import cif
import numpy as np

def ex_info(file, accuracy = 3):
    # Loading file to packages
    mol_1 = MoleculeReader(file)
    doc_1 = cif.read_file(file)
    
    #because of "0.XXXX" is the length of the coordinate of atoms,we need to extend the length by two
    accuracy = accuracy +2
    
    # adding chem_ID and atoms to dataframe
    List = []
    for f in mol_1[0].atoms:
        s = str(f)
        s = s[s.find("(")+1:s.find(")")]
        List.append(s)
    data_frame = pd.DataFrame(data = List, columns = ['atoms'])
    
    # don't know why the output from the atom function is doubled, I need to split this 
    data_frame = data_frame[:int(data_frame.size/2)]

    data_frame['Chem_ID'] = file
    data_frame = data_frame[['Chem_ID','atoms']]
    
    # adding atom number to dataframe
    List =[]
    for f in data_frame['atoms']:
        temp = re.findall(r'\d+', str(f))
        temp_str = int(str(temp[0]))
        List.append(temp_str)
    data_frame['atom#'] = List
    
    # adding 3D coordinate
    block = doc_1.sole_block()
    col_x = block.find_values('_atom_site_fract_x')
    List = []
    for n, x in enumerate(col_x):
        List.append(float((x[0:accuracy])))
    data_frame['X_3D'] = List
    

    col_y = block.find_values('_atom_site_fract_y')
    List = []
    for n, x in enumerate(col_y):
        List.append(float((x[0:accuracy])))
    data_frame['Y_3D'] = List
    
    col_z = block.find_values('_atom_site_fract_z')
    List = []
    for n, x in enumerate(col_z):
        List.append(float((x[0:accuracy])))
    data_frame['Z_3D'] = List
    return data_frame

```
